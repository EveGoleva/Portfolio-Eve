#### Исследование жизненных циклов платформ

Посмотрим сколько лет вообще выпускали игры для каждой из платформ

```python
# Для каждой платформы считаем сколько лет вообще для нее выпускали игры.
# Отдельно выведем последний год выпуска, чтобы понять какие платформы еще актуальны в 2016 году.
#platform_years_sup - platform years of support - годы поддержки платформ

platform_years_sup=data.pivot_table(index='platform',
                                        values=['year_of_release','global_sales',
                                                'jp_sales','eu_sales',
                                                'na_sales','other_sales'],
                                        aggfunc={'year_of_release':[max,min], 
                                                 'global_sales':'sum',
                                                 'jp_sales':'sum',
                                                 'eu_sales':'sum',
                                                 'na_sales':'sum',
                                                 'other_sales':'sum'
                                                })


platform_years_sup.columns = ["_".join((b,a)) for a,b in platform_years_sup.columns]

platform_years_sup=platform_years_sup.sort_values(by='sum_global_sales',ascending=False)
platform_years_sup=platform_years_sup.reset_index()
platform_years_sup['period_of_sales']=platform_years_sup['max_year_of_release']-platform_years_sup['min_year_of_release']
platform_years_sup=platform_years_sup.reindex(columns=['platform','sum_global_sales',
                                                       'sum_eu_sales','sum_jp_sales',
                                                       'sum_na_sales','sum_other_sales','max_year_of_release',
                                                       'min_year_of_release','period_of_sales'])
platform_years_sup.head()
```

```python
platform_years_sup[platform_years_sup['max_year_of_release']!=2016]['period_of_sales'].hist(bins=35);
```

**Из гистограммы видно, что все платформы существовали до 13 лет. Исключение составляет PC, которая является базовой платформой, и DS, которая включает в себя целую линейку платформ.**

```python
#Выведем список платформ, которые поддерживаются в 2016 году.
#X360 - платформа перестала официально поддерживаться в начале 2016 года.
#Wii - платформа перестала официально поддерживаться в 2013 году.
#PS3 - частично поддержка прекращена

actual_platforms=platform_years_sup[platform_years_sup['max_year_of_release']==2016] 
actual_platforms=actual_platforms[(platform_years_sup['platform']!='X360') & 
                                  (platform_years_sup['platform']!='Wii') & 
                                  (platform_years_sup['platform']!='PS3')]
actual_platforms=actual_platforms.sort_values(by='min_year_of_release')
actual_platforms
```

```python
#Группируем продажи по платформе и году релиза
pl_sales_year=data.groupby(['platform','year_of_release'])['global_sales'].sum().reset_index()
# Выберем только актуальные на момент исследования платформы
#act_pl_sales_year - actual platforms sales and years

act_pl_sales_year=pl_sales_year.query('platform in @actual_platforms.platform')
act_pl_sales_year=act_pl_sales_year.groupby(['platform','year_of_release']).agg({'global_sales':'sum'}).reset_index()
#act_pl_sales_year
```

```python
# Построим график, чтобы увидеть жизненный цикл актуальных платформ

sns.set_theme(style="whitegrid", palette=None)

fig,ax=plt.subplots(figsize=(12,6))

sns.lineplot(
    x="year_of_release", 
    y="global_sales",
    data=act_pl_sales_year,
    palette='bright',
    hue='platform',
    );

ax.set(title='Платформы, имеющие официальную поддержку на 2016 год',xlabel='Год выпуска',ylabel='Продажи в мире',xlim=(1985,2015));
```

- PC является базовой платформой - продажи низкие, но постоянные.
- За время существования WiiU и PSV не показали существенных результатов, а также их цикл, основываясь на среднем цикле жизни платформы, подходит к концу, что также показывает падение продаж в последние 2-3 года.
- 3DS настораживает сильное и стабильное падение продаж за последние 5 лет. В реальности это связано с активной поддержкой платформы предыдущего поколения 2DS, которая также считается лучшей по качеству, чем новинка.
- XOne и PS4 - относительно новые платформы, находящиеся в первой фазе своего существования. Если отбросить 2016 год с его неполными данными, то обе платформы переживают хороший рост. Можно предположить, что еще 1-2 года обе платформы будут расти в продажах, затем выйдут на плато на 1-2 года и через 2-4 года будут испытывать спад, пока не будут заменены новыми версиями. **Это платформы, на которые можно делать ставку при проведении маркетинговой кампании.**

### Определение наиболее популярной платформы

Для целей проекта логичнее анализировать популярность актуальных платформ на момент исследования. Так как их всего 6, то Топ-5 делать нецелесообразно.

## Проверка гипотез

```python
# Чтобы избежать ошибок с NaN, отбросим все строки без оценок. Они нам больше не понадобятся.

data_new=data_new.dropna(subset=['user_score'])
```

### Сравнение средних пользовательских рейтингов платформ PC и XOne

Как мы уже увидели, наиболее перспективными новыми платформами в зависимости от региона являются платформы PC и XOne. Сравним их средние пользовательские оценки.

Сформулируем нулевую и альтернативную гипотезы:

- H0- Средний пользовательский рейтинг платформы PS4 равен рейтингу XOne
- H1- Рейтинг двух платформ разный.

Т.к. оценки пользователей это количественные непрерывные величины, то построим гистограмму плотностей распределения для каждой из платформ.

```python
pc=data_new[data_new['platform']=='PC']['user_score']
```

```python
xone=data_new[data_new['platform']=='XOne']['user_score']
```

Т.к. у нас два независимых распределения, необходимо использовать 'st.ttest_ind()'

```python
alpha=0.05

results = st.ttest_ind(
    pc, 
    xone,equal_var=False) # Выборки наверняка разного размера, поэтому рассчитаем дисперсии отдельно

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

H0- Средний пользовательский рейтинг платформы PS4 такой же как у XOne - не может быть отвергнута, хотя рейтинги разные, но с вероятностью 14,8% можно получить такое различие или большее случайно.

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

### Сравнение средних пользовательских рейтингов жанров Action и Sports

Теперь сравним средние пользовательские рейтинги жанров Action и Sports. Нам необходимо проверить разные ли эти значения.

Сформулируем нулевую и альтернативную гипотезы:

- H0- Средний пользовательский рейтинг жанра Action такой же как у Sports
- H1- Средние пользовательские рейтинги двух жанров разные.

```python
action=data_new[data_new['genre']=='Action']['user_score']
```

```python
sports=data_new[data_new['genre']=='Sports']['user_score']
```

Т.к. у нас два независимых распределения, необходимо использовать 'st.ttest_ind()'

```python
alpha=0.05

results = st.ttest_ind(
    action, 
    sports,equal_var=False) # Выборки наверняка разного размера, поэтому рассчитаем дисперсии отдельно

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Вероятность получить результат, что рейтинги жанров Action и Sports равны, случайно слишком мала - 1,44e-20. Поэтому мы вынуждены отвергнуть гипотезу о том, что средние пользовательские рейтинги равны.
